```python
##Data Cleaning
```

```python
df['return_reason'] = df['return_reason'].fillna('No Return')
```

```python
df['coupon_code'] = df['coupon_code'].fillna('No Coupon')
```

```python
df['customer_feedback'] = df['customer_feedback'].fillna('No Feedback')
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['customer_feedback'].value_counts()
```

```python
df[['coupon_code','customer_feedback','return_reason']].head(10)
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df['unit_price_usd'] = df['unit_price_usd'].astype(float)
```

```python
print(df['order_date'].dtype)
```

```python
##Feature Engeneering
```

```python
df['order_frequency'] = df.groupby('order_id')['order_id'].transform('count')
```
